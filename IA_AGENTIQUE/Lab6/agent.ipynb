```python
#Définir une classe de contexte structurée appelée ColourContext
from dataclasses import dataclass
@dataclass
class ColourContext:
    favourite_colour: str = "blue"
    least_favourite_colour: str = "yellow"
```

```python
#Agent sans contexte
from langchain_ollama import ChatOllama
from langchain.agents import create_agent
model = ChatOllama(
model="llama3.2:3b", # ou mistral, gemma, etc.
)
agent = create_agent(model=model,
context_schema=ColourContext)
```

```python
from langchain.messages import HumanMessage
response = agent.invoke(
{"messages": [HumanMessage(content="What is my favourite colour?")]},
context=ColourContext()
)
print(response['messages'][-1].content)
```

```python
#Agent avec contexte
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from langchain.tools import tool, ToolRuntime


@tool
def get_favourite_colour(runtime: ToolRuntime) -> str:
    """
    Get the user's favourite colour.

    Args:
        runtime (ToolRuntime): Runtime context containing user information.

    Returns:
        str: The user's favourite colour.
    """
    return runtime.context.favourite_colour


@tool
def get_least_favourite_colour(runtime: ToolRuntime) -> str:
    """
    Get the user's least favourite colour.

    Args:
        runtime (ToolRuntime): Runtime context containing user information.

    Returns:
        str: The user's least favourite colour.
    """
    return runtime.context.least_favourite_colour


# Create the agent
agent = create_agent(
    model=model,
    tools=[
        get_favourite_colour,
        get_least_favourite_colour,
    ],
    context_schema=ColourContext,
)


# Invoke the agent with a context
response = agent.invoke(
    {
        "messages": [
            HumanMessage(
                content="What is my favourite colour?"
            )
        ]
    },
    context=ColourContext(),
)


# Display the response
print(response["messages"][-1].content)
```

```python
#Changement de contexte
response = agent.invoke(
{"messages": [HumanMessage(content="What is my favourite colour?")]},
context=ColourContext(favourite_colour="green")
)
print(response['messages'][-1].content)
```
